# Dataset Overview
Quick sanity checks for the RecoDAI LUC Scientific Image Forgery Detection data.

```python
from pathlib import Path
import json

DATA_ROOT = (Path('..') / '..' / '..' / 'data' / 'recodai_luc_scientific_image_forgery_detection' / 'raw' /
             'recodai-luc-scientific-image-forgery-detection').resolve()
TRAIN_DIR = DATA_ROOT / 'train_images'
MASK_DIR = DATA_ROOT / 'train_masks'
SAMPLE_OUTPUT = (Path('..') / 'outputs' / 'dataset_overview.json').resolve()

label_counts = {subdir.name: len(list(subdir.glob('*.png'))) for subdir in TRAIN_DIR.iterdir() if subdir.is_dir()}
summary = {"label_counts": label_counts,
           "mask_files": len(list(MASK_DIR.glob('*.npy'))),
           "sample_summary_path": str(SAMPLE_OUTPUT)}
print(json.dumps(summary, indent=2))
```

```python
from PIL import Image
import numpy as np

sample_authentic = sorted((TRAIN_DIR / 'authentic').glob('*.png'))[0]
sample_forged = sorted((TRAIN_DIR / 'forged').glob('*.png'))[0]
sample_mask = sorted(MASK_DIR.glob('*.npy'))[0]

with Image.open(sample_authentic) as img:
    authentic_shape = (img.width, img.height, len(img.getbands()))
with Image.open(sample_forged) as img:
    forged_shape = (img.width, img.height, len(img.getbands()))
mask = np.load(sample_mask)

print('authentic sample:', sample_authentic.name, authentic_shape)
print('forged sample:', sample_forged.name, forged_shape)
print('mask sample:', sample_mask.name, mask.shape, mask.dtype, mask.min(), mask.max())
```
